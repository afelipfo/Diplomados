# 游빑 4.1 NumPy

## 4.1.1 Arreglos (ndarray) y vectores

NumPy proporciona estructuras de datos eficientes como los `ndarray` para trabajar con vectores y matrices.

```python
import numpy as np

# Crear un arreglo unidimensional (vector)
v = np.array([1, 2, 3])
print("Vector:", v)

# Crear una matriz
m = np.array([[1, 2], [3, 4]])
print("Matriz:\n", m)
```

## 4.1.2 Indexaci칩n y slicing en arrays

```python
# Indexado
print("Elemento [0]:", v[0])

# Slicing
print("Subvector:", v[1:])

# Acceso a elementos en matriz
print("Elemento (1,1):", m[1,1])
```

## 4.1.3 Operaciones vectorizadas

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Suma vectorial
print("Suma:", a + b)

# Producto elemento a elemento
print("Producto:", a * b)
```

## 4.1.4 Funciones matem치ticas y estad칤sticas

```python
print("Media:", np.mean(a))
print("Desviaci칩n est치ndar:", np.std(a))
print("Seno:", np.sin(a))
```

## 4.1.5 Broadcasting

Permite realizar operaciones entre arreglos de diferentes formas.

```python
matriz = np.array([[1, 2, 3], [4, 5, 6]])
escalar = 2

# Broadcasting: multiplicar cada elemento por un escalar
print("Multiplicaci칩n:\n", matriz * escalar)
```

## 4.2.1 Series y DataFrames

Pandas permite trabajar con estructuras etiquetadas de datos: `Series` y `DataFrame`.

```python
import pandas as pd

# Series
serie = pd.Series([10, 20, 30], index=["a", "b", "c"])
print("Serie:\n", serie)

# DataFrame
data = {"Nombre": ["Ana", "Luis"], "Edad": [28, 35]}
df = pd.DataFrame(data)
print("\nDataFrame:\n", df)
```

## 4.2.3 Selecci칩n y filtrado (loc, iloc)

```python
print("Con loc:\n", df.loc[0])
print("Con iloc:\n", df.iloc[1])

# Filtrar por condici칩n
print("Mayores de 30 a침os:\n", df[df["Edad"] > 30])
```

## 4.2.4 Agrupamiento (`groupby`) y operaciones estad칤sticas

```python
df2 = pd.DataFrame({
    "Departamento": ["Ventas", "Ventas", "TI", "TI"],
    "Empleado": ["Carlos", "Luc칤a", "Mar칤a", "Pablo"],
    "Salario": [3000, 3200, 4000, 4100]
})

# Agrupar por departamento y calcular media
print(df2.groupby("Departamento")["Salario"].mean())
```

## 4.2.5 Limpieza y transformaci칩n de datos

```python
datos = pd.DataFrame({
    "Nombre": ["Ana", None, "Luis"],
    "Edad": [28, 35, None]
})

# Eliminar filas con valores nulos
print("Sin nulos:\n", datos.dropna())

# Rellenar valores nulos
print("Rellenado:\n", datos.fillna("Desconocido"))

# Cambio de tipos
datos["Edad"] = datos["Edad"].fillna(0).astype(int)
```

# 游늳 4.3 Matplotlib

## 4.3.1 Principios de visualizaci칩n

Matplotlib permite generar gr치ficos est치ticos de alta calidad.

```python
import matplotlib.pyplot as plt

# Datos base
x = [1, 2, 3, 4]
y = [10, 15, 20, 25]
```

## 4.3.2 Gr치ficos de l칤neas, barras, histogramas y scatter

```python
# L칤nea
plt.plot(x, y)
plt.title("Gr치fico de l칤nea")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Barras
plt.bar(["A", "B", "C"], [5, 7, 3])
plt.title("Gr치fico de barras")
plt.show()
```

```python
# Histograma
datos = [1,2,2,3,3,3,4,4,4,4]
plt.hist(datos, bins=4)
plt.title("Histograma")
plt.show()
```

```python
# Dispersi칩n
plt.scatter(x, y)
plt.title("Scatter plot")
plt.show()
```

## 4.3.3 Personalizaci칩n (t칤tulos, ejes, leyendas, colores)

```python
plt.plot(x, y, label="Serie 1", color="green", linestyle="--", marker="o")
plt.title("Personalizaci칩n de gr치fico")
plt.xlabel("Tiempo")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.show()
```

## 4.3.4 Subplots y layouts

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].bar(["A", "B", "C"], [5, 7, 3])
axs[0].set_title("Barras")

axs[1].plot(x, y)
axs[1].set_title("L칤nea")

plt.tight_layout()
plt.show()
```
