## 5.4 Introducci√≥n a pruebas

Las pruebas automatizadas aseguran que tu c√≥digo funciona correctamente y evita errores futuros.

### üì¶ `unittest` (incluido en la librer√≠a est√°ndar)

````python
# test_suma.py
def suma(a, b):
    return a + b

import unittest

class TestSuma(unittest.TestCase):
    def test_positivos(self):
        self.assertEqual(suma(2, 3), 5)

    def test_negativos(self):
        self.assertEqual(suma(-2, -3), -5)

# Ejecutar pruebas
unittest.main(argv=[''], exit=False)
````
