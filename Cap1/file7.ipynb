## **Estad칤stica Inferencial**

La estad칤stica inferencial es una rama de la estad칤stica que se encarga de hacer inferencias sobre las caracter칤sticas de una poblaci칩n a partir de muestras.

# 1. Conceptos B치sicos de Estad칤stica Inferencial

**Poblaci칩n:** Se refiere al conjunto completo de elementos o resultados que se est치n estudiando.
Por ejemplo, si queremos analizar la estatura promedio de todos los estudiantes de una universidad, la poblaci칩n ser칤a todos los estudiantes de esa universidad.

**Muestra:** Es un subconjunto de la poblaci칩n que se selecciona para la observaci칩n y an치lisis.
Continuando con el ejemplo anterior, si no podemos medir la estatura de todos los estudiantes, podr칤amos seleccionar un grupo representativo de 100 estudiantes como nuestra muestra.

**Par치metro:** Un par치metro es una medida num칠rica que describe una caracter칤stica de la poblaci칩n.
En el caso de nuestro ejemplo, la estatura promedio de todos los estudiantes de la universidad ser칤a un par치metro.

**Estad칤stico:** Es una medida num칠rica calculada a partir de los datos de la muestra.
Si calculamos la estatura promedio de nuestra muestra de 100 estudiantes, este promedio ser칤a nuestro estad칤stico.

**Muestreo:** El proceso de seleccionar una muestra de la poblaci칩n. Es crucial que este proceso sea aleatorio para asegurar que la muestra sea representativa de la poblaci칩n.

**Error de muestreo:** La diferencia entre un par치metro poblacional y el estad칤stico calculado a partir de una muestra. Este error surge debido a que la muestra puede no representar perfectamente a toda la poblaci칩n.

Distinguir entre estad칤stica descriptiva e inferencial es fundamental. Mientras que la estad칤stica descriptiva se enfoca en describir y resumir los datos de una muestra o poblaci칩n a trav칠s de medidas como la media, mediana, modo y desviaci칩n est치ndar, la estad칤stica inferencial va m치s all치, utilizando los datos de una muestra para hacer generalizaciones sobre una poblaci칩n. Esto incluye estimar par치metros poblacionales, realizar pruebas de hip칩tesis, y m치s.

Una parte crucial de la estad칤stica inferencial es el muestreo aleatorio. La validez de las inferencias hechas sobre una poblaci칩n depende en gran medida de la representatividad de la muestra. El muestreo aleatorio asegura que cada miembro de la poblaci칩n tenga la misma probabilidad de ser seleccionado para la muestra, minimizando el sesgo y permitiendo que los resultados sean generalizables a toda la poblaci칩n.

```python
import numpy as np
from scipy import stats

# Paso 1: Generar datos de muestra aleatorios
np.random.seed(1000) # Para reproducibilidad
tama침o_muestra = 100
media_poblacional_real = 170
desviacion_estandar_poblacional = 10
muestra = np.random.normal(media_poblacional_real, desviacion_estandar_poblacional, tama침o_muestra)

# Paso 2: Estimaci칩n de la media de la muestra
media_muestra = np.mean(muestra)

(media_muestra)
```

# 2. Distribuciones de Probabilidad


Las distribuciones de probabilidad son fundamentales en la estad칤stica inferencial, ya que nos permiten modelar diferentes tipos de datos y situaciones. Entre las m치s comunes y 칰tiles se encuentran la distribuci칩n normal, la t-Student, la chi-cuadrado, y la F, cada una con sus propios usos y caracter칤sticas en el contexto de la inferencia estad칤stica.

**Distribuci칩n Normal:** Conocida tambi칠n como distribuci칩n gaussiana, es quiz치s la m치s importante y ampliamente utilizada debido a su propiedad fundamental descrita en el Teorema del L칤mite Central. Este teorema establece que, bajo ciertas condiciones, la suma (o promedio) de un gran n칰mero de variables aleatorias independientes y distribuidas id칠nticamente tiende a seguir una distribuci칩n normal, independientemente de la distribuci칩n original de las variables. Esto hace que la distribuci칩n normal sea central en la estad칤stica inferencial, pues muchas t칠cnicas asumen que los datos se distribuyen de esta manera, especialmente cuando el tama침o de la muestra es grande.

**Distribuci칩n t-Student:** Esta distribuci칩n es particularmente 칰til cuando se trabaja con muestras peque침as y la desviaci칩n est치ndar de la poblaci칩n es desconocida. La t-Student ayuda a ajustar la incertidumbre adicional en la estimaci칩n del error est치ndar y se utiliza ampliamente en pruebas de hip칩tesis y en la construcci칩n de intervalos de confianza cuando el tama침o de la muestra es reducido.

**Distribuci칩n Chi-cuadrado (픦):** Se aplica principalmente en pruebas de bondad de ajuste, pruebas de independencia en tablas de contingencia y en la estimaci칩n de varianzas. Una caracter칤stica clave de la distribuci칩n chi-cuadrado es que se utiliza para analizar variables aleatorias que deben ser siempre positivas, ya que se basa en sumas de cuadrados de variables aleatorias normales estandarizadas.

**Distribuci칩n F:** Se utiliza en el an치lisis de varianza (ANOVA), para comparar las varianzas de dos poblaciones y en el an치lisis de regresi칩n. La distribuci칩n F es clave para determinar si las diferencias entre las medias de diferentes grupos son estad칤sticamente significativas, permitiendo a los investigadores evaluar la varianza entre grupos frente a la varianza dentro de los grupos.

```python
import numpy as np
from scipy import stats

# 1. Distribuci칩n Normal: Generaci칩n de datos
datos_normales = np.random.normal(loc=0, scale=1, size=1000)  # media=0, desviaci칩n est치ndar=1

# 2. Distribuci칩n t-Student: Intervalo de confianza para la media de una muestra peque침a
muestra_peque침a = np.random.choice(datos_normales, size=30)  # Tomamos una muestra de 30 elementos
confianza = 0.95
grados_libertad = len(muestra_peque침a)-1
media_muestra = np.mean(muestra_peque침a)
sem = stats.sem(muestra_peque침a)  # Error est치ndar de la media
intervalo_confianza = stats.t.interval(confianza, grados_libertad, media_muestra, sem)

# 3. Distribuci칩n Chi-cuadrado: Prueba de bondad de ajuste
# Supongamos que queremos comprobar si nuestra 'datos_normales' sigue una distribuci칩n normal
# Para simplificar, usaremos la prueba de kurtosis, que es una parte de lo que podr칤a implicar una prueba de bondad de ajuste
kurtosis_test = stats.kurtosistest(datos_normales)

# 4. Distribuci칩n F: ANOVA simplificado para comparar varianzas
# Generamos dos grupos de datos para comparar sus varianzas
datos_grupo1 = np.random.normal(loc=0, scale=1, size=30)
datos_grupo2 = np.random.normal(loc=0, scale=1.5, size=30)
#anova_resultado = stats.f_oneway(datos_grupo1, datos_grupo2)

(intervalo_confianza, kurtosis_test) #anova_resultado
```

# 3. Estimaci칩n de Par치metros

La estimaci칩n de par치metros poblacionales es un proceso crucial en estad칤stica inferencial que nos permite hacer generalizaciones sobre una poblaci칩n entera a partir del an치lisis de una muestra de dicha poblaci칩n. Los par치metros poblacionales, como la media y la varianza, son valores que resumen caracter칤sticas de la totalidad de la poblaci칩n. Dado que no siempre es pr치ctico o posible estudiar cada miembro de una poblaci칩n, utilizamos muestras y a partir de ellas calculamos estad칤sticos muestrales, como la media y la varianza de la muestra, para estimar dichos par치metros.

Existen dos tipos principales de estimaciones: puntuales e intervalares. Una estimaci칩n puntual proporciona un 칰nico valor como estimaci칩n del par치metro poblacional. Por ejemplo, si queremos estimar la altura media de todos los estudiantes de una universidad, podr칤amos tomar una muestra de estudiantes, calcular la media de la altura de esos estudiantes (estad칤stico muestral), y usar ese valor como nuestra estimaci칩n puntual de la altura media de la poblaci칩n de todos los estudiantes.

Sin embargo, las estimaciones puntuales tienen una limitaci칩n importante: no nos dicen nada sobre la precisi칩n o la incertidumbre asociada con la estimaci칩n. Aqu칤 es donde entran en juego las estimaciones por intervalo, que proporcionan un rango de valores dentro del cual esperamos que se encuentre el par치metro poblacional. Este rango se conoce como intervalo de confianza.

El intervalo de confianza es una herramienta poderosa en estad칤stica inferencial porque no solo ofrece una estimaci칩n del par치metro poblacional, sino que tambi칠n indica un nivel de confianza en esa estimaci칩n. Por ejemplo, un intervalo de confianza del 95% para la altura media de los estudiantes significa que, si repiti칠ramos el proceso de muestreo muchas veces, esperar칤amos que el par치metro poblacional real (la altura media de todos los estudiantes) estuviera dentro de ese intervalo en el 95% de los casos.

La importancia del intervalo de confianza radica en su capacidad para expresar la precisi칩n de una estimaci칩n. Un intervalo m치s estrecho indica una mayor precisi칩n y un alto nivel de confianza en que el par치metro poblacional se encuentra dentro de ese intervalo. Por el contrario, un intervalo m치s amplio sugiere mayor incertidumbre sobre la ubicaci칩n exacta del par치metro. La selecci칩n del nivel de confianza (como 95% o 99%) y el c치lculo del intervalo de confianza dependen del tama침o de la muestra y de la variabilidad de los datos en la muestra.

```python
# Importamos las librer칤as necesarias
import numpy as np

#Importando Scipy
from scipy import stats

# Generamos una muestra aleatoria para el ejemplo
# Supongamos que esta muestra representa las alturas (en cm) de un grupo de estudiantes
np.random.seed(42)  # Para garantizar reproducibilidad
muestra_alturas = np.random.normal(170, 10, 100)  # media = 170 cm, desviaci칩n est치ndar = 10 cm, n = 100

# Estimaci칩n puntual de la media y la varianza de la muestra
media_muestra = np.mean(muestra_alturas)
varianza_muestra = np.var(muestra_alturas, ddof=1)  # ddof=1 para la varianza muestral

# Estimaci칩n por intervalo (Intervalo de confianza) para la media
# Usamos un intervalo de confianza del 95%
confianza = 0.95
sem = stats.sem(muestra_alturas)  # Error est치ndar de la media
intervalo_confianza = stats.t.interval(confianza, len(muestra_alturas)-1, media_muestra, sem)

(media_muestra, varianza_muestra, intervalo_confianza)
```

# 4. Pruebas de Hip칩tesis
Resumen: Se introducen los conceptos de hip칩tesis nula y alternativa, y se explica c칩mo realizar pruebas de hip칩tesis para evaluar afirmaciones sobre par치metros poblacionales. Se discuten los errores tipo I y tipo II, el valor p y el nivel de significancia.

Las pruebas de hip칩tesis son un pilar fundamental de la estad칤stica inferencial, proporcionando un marco formal para tomar decisiones y evaluar afirmaciones sobre par치metros poblacionales basados en datos de muestra. Este proceso comienza con la formulaci칩n de dos hip칩tesis opuestas: la hip칩tesis nula (H0) y la hip칩tesis alternativa (H1 o Ha).

Hip칩tesis Nula (H0): Es una declaraci칩n de "no efecto" o "no diferencia" que indica que el par치metro poblacional es igual a un valor especificado. La hip칩tesis nula es lo que se pone a prueba en un an치lisis estad칤stico, y el objetivo es determinar si hay suficiente evidencia en los datos de la muestra para rechazarla.

Hip칩tesis Alternativa (H1 o Ha): Contrasta directamente con la hip칩tesis nula y representa una afirmaci칩n de que hay un efecto, una diferencia o una relaci칩n. La hip칩tesis alternativa se acepta si los datos de la muestra proporcionan suficiente evidencia para rechazar la hip칩tesis nula.

Errores Tipo I y Tipo II:

Error Tipo I (풤): Ocurre cuando la hip칩tesis nula es rechazada incorrectamente, aunque en realidad es verdadera. El nivel de significancia (풤), com칰nmente fijado en 0.05, indica la probabilidad de cometer un error de Tipo I.
Error Tipo II (): Se da cuando la hip칩tesis nula no es rechazada, a pesar de que la hip칩tesis alternativa es verdadera. La potencia de una prueba (1 - ) mide la capacidad de la prueba para detectar un efecto real cuando existe.
Valor P: Es la probabilidad de observar los resultados de la muestra, o algo m치s extremo, asumiendo que la hip칩tesis nula es cierta. Un valor p bajo sugiere que los datos observados son inusuales bajo la suposici칩n de que la hip칩tesis nula es verdadera. Si el valor p es menor que el nivel de significancia 풤, se rechaza la hip칩tesis nula, indicando que hay evidencia suficiente para apoyar la hip칩tesis alternativa.

Nivel de Significancia (풤): Es el umbral utilizado para determinar si el valor p es lo suficientemente bajo como para rechazar la hip칩tesis nula. Un nivel de significancia com칰n es 0.05, aunque este valor puede ajustarse dependiendo del contexto y la tolerancia al riesgo de cometer un error de Tipo I.

El proceso de realizar una prueba de hip칩tesis implica seleccionar la prueba estad칤stica adecuada para los datos y los objetivos del estudio, calcular el estad칤stico de prueba y el valor p correspondiente, y luego comparar el valor p con el nivel de significancia para tomar una decisi칩n sobre las hip칩tesis.

```python
# Para ilustrar una prueba de hip칩tesis, vamos a utilizar un ejemplo sencillo:
# Supongamos que queremos probar si la media de una muestra difiere de un valor poblacional conocido.
# Por ejemplo, queremos saber si la altura media de un grupo de estudiantes es diferente de 170 cm.

# Generamos datos de muestra aleatorios que representen las alturas de los estudiantes
np.random.seed(0)  # Para garantizar la reproducibilidad
alturas_muestra = np.random.normal(172, 5, 30)  # media = 172 cm, desviaci칩n est치ndar = 5 cm, n = 30

# Definimos la hip칩tesis nula y la hip칩tesis alternativa:
# H0: La media de la altura es 170 cm (풮 = 170)
# H1: La media de la altura no es 170 cm (풮  170)

# Nivel de significancia
alpha = 0.05

# Realizamos una prueba t de una muestra para comparar la media muestral con la media poblacional conocida de 170 cm
t_stat, p_valor = stats.ttest_1samp(alturas_muestra, 170)

# Decisi칩n basada en el valor p y el nivel de significancia
resultado = "rechazar la hip칩tesis nula" if p_valor < alpha else "no rechazar la hip칩tesis nula"

(t_stat, p_valor, resultado)
```

# 5. Regresi칩n y Correlaci칩n
Resumen: Se examina c칩mo modelar la relaci칩n entre dos o m치s variables mediante la regresi칩n lineal y no lineal, y c칩mo medir la fuerza y direcci칩n de la relaci칩n con el coeficiente de correlaci칩n. Se discuten los conceptos de causalidad y correlaci칩n.

En el estudio de la estad칤stica, los conceptos de regresi칩n y correlaci칩n son fundamentales para analizar y entender la relaci칩n entre variables. Estas herramientas nos permiten examinar c칩mo una variable puede predecir o estar asociada con otra, proporcionando una base s칩lida para la toma de decisiones basada en datos y el an치lisis cient칤fico.

Regresi칩n: La regresi칩n se centra en modelar y analizar la relaci칩n entre una variable dependiente (o variable de respuesta) y una o m치s variables independientes (o predictores). El objetivo es entender c칩mo la variable dependiente cambia con respecto a las variables independientes. La regresi칩n puede ser lineal o no lineal.

Regresi칩n Lineal: En la regresi칩n lineal, se asume que existe una relaci칩n lineal entre las variables. Esto se representa mediante una l칤nea recta en un gr치fico bidimensional, donde una variable es explicada o predicha por otra. La regresi칩n lineal es adecuada cuando la relaci칩n entre las variables se puede aproximar por una l칤nea recta.
Regresi칩n No Lineal: Cuando la relaci칩n entre las variables no sigue una l칤nea recta, se utiliza la regresi칩n no lineal. Este tipo de regresi칩n es 칰til para modelar relaciones m치s complejas, donde los cambios en una variable independiente pueden tener efectos no proporcionales en la variable dependiente.
Correlaci칩n: Mientras que la regresi칩n proporciona un modelo para la relaci칩n entre variables, la correlaci칩n mide la fuerza y la direcci칩n de esta relaci칩n. El coeficiente de correlaci칩n var칤a entre -1 y 1, donde valores cercanos a 1 o -1 indican una fuerte relaci칩n lineal positiva o negativa, respectivamente, y valores cercanos a 0 indican una falta de relaci칩n lineal.

Correlaci칩n Positiva: Ocurre cuando ambas variables se mueven en la misma direcci칩n; es decir, cuando una variable aumenta, la otra tambi칠n lo hace.
Correlaci칩n Negativa: Se da cuando las variables se mueven en direcciones opuestas; es decir, cuando una variable aumenta, la otra disminuye.
Es crucial distinguir entre correlaci칩n y causalidad. Mientras que la correlaci칩n indica la existencia de una relaci칩n entre dos variables, no implica que una variable cause el cambio en la otra. La causalidad, por otro lado, implica que un cambio en una variable provoca un cambio en otra. Establecer causalidad requiere un an치lisis m치s profundo y, a menudo, experimentaci칩n controlada o t칠cnicas estad칤sticas avanzadas para descartar factores confundentes.

```python
# Para ilustrar regresi칩n y correlaci칩n, usaremos un conjunto de datos ficticios que representan
# las horas de estudio (variable independiente) y las calificaciones obtenidas (variable dependiente).

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress

# Generamos datos ficticios
np.random.seed(1)  # Para garantizar la reproducibilidad
horas_estudio = np.random.uniform(0, 10, 50)  # Horas de estudio var칤an de 0 a 10
calificaciones = 5 + horas_estudio * 3 + np.random.normal(0, 2, 50)  # Calificaci칩n basada en horas de estudio m치s ruido

# Creamos un DataFrame con pandas
datos = pd.DataFrame({
    'Horas de Estudio': horas_estudio,
    'Calificaciones': calificaciones
})

# Calculamos la correlaci칩n
coef_correlacion = datos.corr()

# Realizamos una regresi칩n lineal
slope, intercept, r_value, p_value, std_err = linregress(horas_estudio, calificaciones)

# Predecimos calificaciones usando la regresi칩n lineal para graficar
calificaciones_predichas = intercept + slope * horas_estudio

# Graficamos los datos junto con la l칤nea de regresi칩n
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Horas de Estudio', y='Calificaciones', data=datos, label='Datos')
plt.plot(horas_estudio, calificaciones_predichas, color='red', label='L칤nea de Regresi칩n')
plt.title('Regresi칩n Lineal: Horas de Estudio vs. Calificaciones')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.legend()
plt.show()

(coef_correlacion, r_value**2)
```

## **Distribuci칩n Binomial**

### **Enunciado del problema**

El 80% de los visitantes del Museo Lavista terminan comprando recuerdos en la tienda de souvenirs del museo. Este domingo, si se selecciona una muestra aleatoria de 10 visitantes:

1. Encuentra la **probabilidad de que todos los visitantes** terminen comprando en la tienda de souvenirs.  
2. Encuentra la **probabilidad de que como m치ximo 7 visitantes** compren recuerdos en la tienda de souvenirs.

---

### **Verificaci칩n de supuestos de la distribuci칩n binomial**

- Solo hay **dos resultados posibles** (칠xito o fracaso) para cada intento: un visitante comprar치 (s칤) o no comprar치 (no) souvenirs.
- El **n칰mero de intentos (n)** es fijo: hay 10 visitantes en la muestra.
- Cada intento es **independiente** de los dem치s: es razonable asumir que la decisi칩n de compra de cada visitante es independiente.
- La **probabilidad de 칠xito (p)** es la misma en cada intento: la probabilidad de que un visitante compre es **0.8**.

```python
# Importar los paquetes importantes
import pandas as pd  # Librer칤a utilizada para manipulaci칩n y an치lisis de datos
import numpy as np  # Librer칤a utilizada para trabajar con arreglos (arrays)
import matplotlib.pyplot as plt  # Librer칤a para gr치ficos y visualizaciones
import seaborn as sns  # Librer칤a para visualizaciones
%matplotlib inline

import scipy.stats as stats  # Esta librer칤a contiene una gran cantidad de distribuciones de probabilidad y funciones estad칤sticas

# Para suprimir advertencias
import warnings
warnings.filterwarnings('ignore')
```

### **Estimemos la distribuci칩n de probabilidad de los visitantes**

A continuaci칩n, declaramos el tama침o de la muestra y la probabilidad de 칠xito. Adem치s, generamos los posibles valores de visitantes que compran souvenirs, desde 0 hasta 10.

```python
# Declarar el tama침o de la muestra, donde n representa el n칰mero de visitantes seleccionados aleatoriamente
n = 10

# Declarar p, que representa la probabilidad de 칠xito, es decir, que un visitante compre un souvenir
p = 0.80

# Declarar los diferentes posibles n칰meros de visitantes que compran, usando un arreglo de numpy
k = np.arange(0, 11)
k
```

### **Funci칩n de probabilidad de la distribuci칩n binomial**

La funci칩n de probabilidad de la distribuci칩n binomial proporciona la probabilidad de obtener **x 칠xitos** en **n intentos**, donde **p** es la probabilidad de 칠xito:

>$P(X = x) = {n \choose x} p^x (1 - p)^{n - x}$

En este caso, conocemos:

- **n** (n칰mero de visitantes seleccionados aleatoriamente) = 10  
- **p** (probabilidad de 칠xito, es decir, que un visitante compre un souvenir) = 0.80  
- **q** (probabilidad de fracaso, es decir, que un visitante **no** compre un souvenir) = 1 - 0.80 = 0.20  
- **x** (n칰mero de 칠xitos) = 10  

Usaremos la funci칩n `binom.pmf()` para calcular esta distribuci칩n de probabilidad, que nos dar치 la probabilidad de que un n칰mero determinado de visitantes (de un total de 10) termine comprando en la tienda de souvenirs.

```python
# Importar la funci칩n necesaria
from scipy.stats import binom

# Utilizar la funci칩n binom.pmf() para generar la distribuci칩n de probabilidad
binomial = binom.pmf(k=k, n=n, p=p)

binomial
```

Como puedes ver, **binomial** es un arreglo de probabilidades correspondientes a los diferentes posibles n칰meros de 칠xitos. Vamos a visualizar esta distribuci칩n con el siguiente gr치fico.

```python
# Graficar la distribuci칩n
plt.bar(k, binomial)  # Crear un gr치fico de barras
plt.title("Binomial: n=%i , p=%.2f" % (n, p), fontsize=15)  # T칤tulo del gr치fico
plt.xlabel("N칰mero de 칠xitos")  # Etiqueta del eje x
plt.ylabel("Probabilidad de 칠xitos")  # Etiqueta del eje y
plt.show()  # Mostrar el gr치fico
```

El 칰ltimo elemento del arreglo **binomial** representa la probabilidad de que el n칰mero de 칠xitos sea **10**, es decir, que **los 10 visitantes seleccionados aleatoriamente compren souvenirs** en la tienda.

La siguiente instrucci칩n nos da esa probabilidad:

```python
binomial[10]
```

La probabilidad obtenida es aproximadamente **0.11**, lo que indica que hay un 11% de probabilidad de que **todos** los visitantes compren souvenirs.

---

Ahora nos interesa calcular la **probabilidad de que como m치ximo 7 visitantes** (de los 10 seleccionados aleatoriamente) compren souvenirs, es decir:

> \( P(X \leq 7) \)

Para ello, utilizamos la **CDF (Funci칩n de Distribuci칩n Acumulada)**, la cual representa la probabilidad acumulada de que la variable aleatoria \( X \) tome un valor menor o igual a un valor dado \( x \):

> \( F_X(x) = P(X \leq x) \)

En este caso, \( X \) representa el n칰mero de visitantes que compran souvenirs.

A continuaci칩n, mostramos una visualizaci칩n opcional donde se resaltan en rojo las probabilidades acumuladas hasta 7 칠xitos:

```python
# Graficar la distribuci칩n de probabilidad
barl = plt.bar(k, binomial)  # Crear gr치fico de barras
plt.title("Binomial: n=%i , p=%.2f" % (n, p), fontsize=15)  # T칤tulo
plt.xlabel("N칰mero de 칠xitos")  # Etiqueta del eje X
plt.ylabel("Probabilidad de 칠xitos")  # Etiqueta del eje Y

# Resaltar en rojo las barras donde X <= 7
for i in range(0, 8):
    barl[i].set_color("r")

plt.show()  # Mostrar el gr치fico
```

En el gr치fico anterior, la regi칩n roja representa \( P(X \leq 7) \).

Ahora vamos a **calcular esa probabilidad** usando la funci칩n `binom.cdf()` de SciPy, que retorna la probabilidad acumulada de obtener **como m치ximo** 7 칠xitos (es decir, que hasta 7 visitantes compren souvenirs).

```python
# Calcular la CDF para X <= 7
binom.cdf(k=7, n=n, p=p)
```

### **Conclusi칩n**

- Hay una probabilidad de **10.74%** de que **todos los visitantes** terminen comprando recuerdos en la tienda del museo.
- Hay una probabilidad de **32.22%** de que **como m치ximo 7 visitantes (de 10)** compren souvenirs en la tienda del museo.

Vamos a modificar la **probabilidad de 칠xito** (es decir, la probabilidad de que cada visitante compre souvenirs) a distintos valores: **60%**, **70%**, **90%** y **95%**, para visualizar c칩mo cambia la forma de la distribuci칩n binomial.

A continuaci칩n se muestra un conjunto de gr치ficos que comparan estas distribuciones para un total de 10 visitantes:

```python
plt.figure(figsize=(15, 4))

# Distribuci칩n para p = 0.70
plt.subplot(131)
binomial_70 = binom.pmf(k, n, p=0.7)
plt.bar(k, binomial_70)
plt.title("p=%.2f" % (0.7), fontsize=15)
plt.xlabel("N칰mero de 칠xitos")
plt.ylabel("Probabilidad de 칠xitos")

# Distribuci칩n para p = 0.80
plt.subplot(132)
binomial_80 = binom.pmf(k, n, p=0.8)
plt.bar(k, binomial_80)
plt.title("p=%.2f" % (0.8), fontsize=15)
plt.xlabel("N칰mero de 칠xitos")
plt.ylabel("Probabilidad de 칠xitos")

# Distribuci칩n para p = 0.90
plt.subplot(133)
binomial_90 = binom.pmf(k, n, p=0.9)
plt.bar(k, binomial_90)
plt.title("p=%.2f" % (0.9), fontsize=15)
plt.xlabel("N칰mero de 칠xitos")
plt.ylabel("Probabilidad de 칠xitos")

plt.tight_layout(w_pad=5)
plt.show()
```

### **Conclusi칩n**

Es evidente en los gr치ficos anteriores que **la forma de la distribuci칩n cambia a medida que cambia el valor de \( p \)** (la probabilidad de 칠xito):

- Cuando \( p \) es **menor**, la distribuci칩n est치 m치s **sim칠trica** y centrada hacia la izquierda.
- A medida que \( p \) **aumenta**, la distribuci칩n se **desplaza hacia la derecha**, concentrando m치s probabilidad en los valores cercanos al n칰mero m치ximo de 칠xitos (n=10).

Esto refleja que **a mayor probabilidad de 칠xito**, es m치s probable observar un mayor n칰mero de 칠xitos en los ensayos.

## **Teorema Central del L칤mite (TCL)**

El **Teorema Central del L칤mite (TCL)** establece que si tomamos m칰ltiples muestras independientes de una poblaci칩n, calculamos la **media de cada muestra** y graficamos estas medias, entonces esa distribuci칩n tender치 a una **distribuci칩n normal** a medida que el tama침o de las muestras aumente, **sin importar** la forma de la distribuci칩n original de la poblaci칩n.

---

### Veamos el TCL en acci칩n con una simulaci칩n en Python

#### A continuaci칩n, generamos una distribuci칩n **Uniforme** (que definitivamente **no es Normal**)

### **Importar y visualizar la poblaci칩n con distribuci칩n uniforme**

Primero generamos una **poblaci칩n de datos distribuidos uniformemente** entre 0 y 10. Esta distribuci칩n no tiene forma normal.

A continuaci칩n, visualizamos dicha poblaci칩n:

```python
# Importar la funci칩n necesaria
from scipy.stats import uniform

# Fijar la semilla para reproducibilidad
np.random.seed(1)

# Crear una poblaci칩n con distribuci칩n uniforme de tama침o 100,000
uniform_pop = uniform.rvs(0, 10, size=100000)

# Visualizar la distribuci칩n uniforme
plt.hist(uniform_pop)
plt.title("Poblaci칩n con Distribuci칩n Uniforme")
plt.xlabel("X ~ U(0,10)")
plt.ylabel("Frecuencia")
plt.show()
```

---

### **Creaci칩n de una distribuci칩n muestral**  
Vamos a crear una **distribuci칩n de medias muestrales** a partir de esta poblaci칩n uniforme.

- Extraeremos una muestra de tama침o **n = 5**
- Calcularemos la **media** de esa muestra
- Repetiremos este proceso **500 veces**, obteniendo as칤 **500 medias muestrales**

Esto nos permitir치 observar c칩mo **el Teorema Central del L칤mite** act칰a incluso cuando la poblaci칩n **no es normal**.

Ahora veamos c칩mo se comporta esta distribuci칩n muestral:

```python
# Fijar la semilla para reproducibilidad
np.random.seed(1)

# Establecer tama침o de muestra
n = 5

# Lista para guardar las medias muestrales
sample_means = []

# Repetir el proceso 500 veces
for j in range(500):
    # Tomar una muestra aleatoria de tama침o n desde la poblaci칩n uniforme
    sample = np.random.choice(uniform_pop, size=n)
    
    # Calcular la media de la muestra
    sample_mean = np.mean(sample)
    
    # Almacenar la media en la lista
    sample_means.append(sample_mean)

# Graficar el histograma de las medias muestrales
sns.displot(sample_means, kde=True)
plt.title("Distribuci칩n de Medias Muestrales para n = " + str(n))
plt.xlabel("Media muestral")
plt.ylabel("Frecuencia")
plt.show()
```

#### **Generemos otra distribuci칩n muestral aumentando el tama침o de muestra a 15 (n = 15)**

Vamos a repetir el experimento anterior pero con un tama침o de muestra mayor. Seg칰n el Teorema Central del L칤mite, **al aumentar el tama침o de muestra**, la distribuci칩n de las medias muestrales se aproxima a칰n m치s a una **distribuci칩n normal**.

```python
# Fijar la semilla para reproducibilidad
np.random.seed(1)

# Establecer tama침o de muestra
n = 15

# Lista para almacenar medias muestrales
sample_means = []

# Repetir el experimento 500 veces
for j in range(500):
    sample = np.random.choice(uniform_pop, size=n)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# Graficar la distribuci칩n de las medias
sns.displot(sample_means, kde=True)
plt.title("Distribuci칩n de Medias Muestrales para n = " + str(n))
plt.xlabel("Media muestral")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Fijar la semilla para reproducibilidad
np.random.seed(1)

# Establecer tama침o de muestra
n = 30

# Lista para almacenar medias muestrales
sample_means = []

# Repetir el experimento 500 veces
for j in range(500):
    sample = np.random.choice(uniform_pop, size=n)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# Graficar la distribuci칩n de las medias
sns.displot(sample_means, kde=True)
plt.title("Distribuci칩n de Medias Muestrales para n = " + str(n))
plt.xlabel("Media muestral")
plt.ylabel("Frecuencia")
plt.show()
```

#### **Aumentemos ahora el tama침o de muestra a 50 (n = 50)**

```python
# Fijar la semilla para reproducibilidad
np.random.seed(1)

# Establecer tama침o de muestra
n = 50

# Lista para almacenar medias muestrales
sample_means = []

# Repetir el experimento 500 veces
for j in range(500):
    sample = np.random.choice(uniform_pop, size=n)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# Graficar la distribuci칩n de las medias
sns.displot(sample_means, kde=True)
plt.title("Distribuci칩n de Medias Muestrales para n = " + str(n))
plt.xlabel("Media muestral")
plt.ylabel("Frecuencia")
plt.show()
```

游댌 A medida que el **tama침o de la muestra aumenta**, la **distribuci칩n de las medias muestrales** se vuelve **m치s parecida a una distribuci칩n normal**.

#### **쮽unciona el Teorema Central del L칤mite con una poblaci칩n Exponencial?**

## **Intervalo de Confianza**

### **쮺칩mo se construye un intervalo de confianza para la media poblacional cuando la desviaci칩n est치ndar es conocida?**

Es poco com칰n conocer la desviaci칩n est치ndar de la poblaci칩n y no la media, pero en procesos industriales muy controlados, esto puede suceder.  
Por ejemplo, la variabilidad puede ser estable en el tiempo, aunque peque침os cambios (temperatura, humedad) afecten la media.

Veamos un ejemplo:

> El contenido de cafe칤na (en mg) fue examinado para una muestra aleatoria de **50 tazas** de caf칠 negro servidas por una nueva m치quina.  
> La **media muestral** fue de **110 mg**.  
> Se sabe que la **desviaci칩n est치ndar poblacional** para este tipo de m치quinas es de **7 mg**.  
> Construye un intervalo de confianza del **95%** para \( \mu \), el contenido medio de cafe칤na.

---

```python
# Importar funci칩n necesaria
from scipy.stats import norm

# Media muestral y desviaci칩n est치ndar poblacional
x_bar, sigma = 110, 7

# Tama침o de la muestra
n = 50

# Calcular el intervalo de confianza al 95%
np.round(norm.interval(0.95, loc=x_bar, scale=sigma / np.sqrt(n)), 2)
```

### 游 Insight

九 El **95% de las veces**, el contenido medio de cafe칤na estar치 entre **108.06 mg** y **111.94 mg**.

---

### **쯈u칠 pasa si no se conoce la desviaci칩n est치ndar poblacional?**

En la mayor칤a de los casos reales, **no se conoce la desviaci칩n est치ndar poblacional**.  
En ese caso, usamos la **desviaci칩n est치ndar muestral** y el **intervalo se construye con la distribuci칩n t de Student** con \( n - 1 \) grados de libertad.

游늷 La distribuci칩n t:

- Es sim칠trica y centrada en cero.
- Se parece a la normal cuando los grados de libertad son grandes.

```python
# Importar funciones necesarias
from scipy.stats import t, norm

# Valores del eje x
x = np.linspace(-3, 3, 100)

# Graficar la distribuci칩n t para diferentes grados de libertad
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()
for i, k in zip(range(6), [1, 2, 3, 5, 10, 30]):
    ax = axes[i]
    ax.plot(x, t.pdf(x, df=k), color="blue", label="Distribuci칩n t")
    ax.plot(x, norm.pdf(x), color="red", label="Distribuci칩n normal")
    ax.set_title(f"Distribuci칩n t con k={k}")
    ax.legend(loc="upper right", fontsize=10)
plt.tight_layout()
```

### **Construyamos el intervalo de confianza con desviaci칩n est치ndar desconocida**

> Se examinaron 50 tazas.  
> La **media muestral** fue 110 mg y la **desviaci칩n est치ndar muestral** fue 7 mg.  
> Construye un intervalo de confianza del 95% para \( \mu \).

---


```python
# Importar funci칩n t
from scipy.stats import t

# Media muestral y desviaci칩n est치ndar muestral
x_bar, s = 110, 7

# Tama침o de la muestra y grados de libertad
n = 50
k = n - 1

# Calcular el intervalo de confianza con distribuci칩n t
np.round(t.interval(0.95, df=k, loc=x_bar, scale=s / np.sqrt(n)), 2)
```

### 游 Insight
