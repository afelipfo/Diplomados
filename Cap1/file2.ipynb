# Manejo de Errores

---

El manejo de errores en Python se refiere a la implementaci贸n de estructuras y t茅cnicas que permiten gestionar y controlar las excepciones o errores que pueden surgir durante la ejecuci贸n de un programa. Las excepciones son eventos inesperados o condiciones an贸malas que pueden ocurrir durante la ejecuci贸n del c贸digo y que pueden interrumpir el flujo normal del programa si no se gestionan adecuadamente. Para lidiar con posibles excepciones en el programa se hace uso de las palabras reservadas `try-except`.

La sintaxis b谩sica de un `try-except` es

```python
# Sintaxis b谩sica del manejo de errores o excepciones

try:
    # C贸digo propenso a generar excepciones
except ExcepcionTipo1 as e1:
    # Manejar excepci贸n de tipo 1
except ExcepcionTipo2 as e2:
    # Manejar excepci贸n de tipo 2
```

Opcionalmente, se pueden utilizar bloques `else` y `finally` despu茅s de los bloques except. El bloque `else` se ejecuta si no se produce ninguna excepci贸n en el bloque `try`. El bloque `finally` se ejecuta siempre, independientemente de si se produjo una excepci贸n o no.

```python
# Sintaxis b谩sica del manejo de errores o excepciones con bloques else y finally

try:
    # C贸digo propenso a generar excepciones
except ExcepcionTipo1 as e1:
    # Manejar excepci贸n de tipo 1
except ExcepcionTipo2 as e2:
    # Manejar excepci贸n de tipo 2
else:
    # C贸digo a ejecutar si no hay excepciones
finally:
    # C贸digo que siempre se ejecutar谩
```

```python
# Ejemplo simple de try-except

a = 13

try:
    b = int(input("Ingrese un numero para dividir a 'a' =  "))
    a = a/b
    print("Su resultado es =",a)
except:
    print("Hay un error!")
```

Un error o excepcion puede ser "bloqueado" mediante una o varias l铆neas de c贸digo. Tamb铆en, los bloques `except` pueden ser programadas para tener en cuenta alg煤n tipo de excepci贸n espec铆fica o no.

```python
# Ejemplo de m煤ltiples excepciones dentro de un m铆smo bloque except

try:
    # C贸digo propenso a generar excepciones
except (ZeroDivisionError, NameError):
    # C贸digo a ejecutar si hay excepciones ZeroDivisionError ni NameError

# codigo a ejecutar si no ex铆sten excepciones
```

```python
# Ejemplo de m煤ltiples excepciones programadas expl铆citamente

try:
    # C贸digo propenso a generar excepciones
except ZeroDivisionError:
    # C贸digo a ejecutar si ex铆ste un ZeroDivisionError
except NameError:
    # C贸digo a ejecutar si ex铆ste un NameError
except:
    # C贸digo a ejecutar si ex铆ste alguna excepcion en general

# codigo a ejecutar si no ex铆sten excepciones
```

Por ejemplo, podr茅mos ejecutar diferentes bloques de c贸digo dependiendo de casos espec铆ficos.

```python
# Ejemplo de manejo de excepciones/errores espec铆ficas

a = 1

try:
    b = int(input("Ingrese un n煤mero para dividir a 'a' = "))
    a = a/b
    print("Resultado = ",a)
except ZeroDivisionError:
    print("No es posible realizar la divisi贸n ya que usted ingres贸 a = 0")
except ValueError:
    print("Usted no ingres贸 n煤mero alguno.")
except:
    print("Algo sali贸 mal, que puede ser?")
```

```python
# Ejemplo de manejo de excepciones/errores espec铆ficas con bloque finally

a = 1

try:
    b = int(input("Ingrese un n煤mero para dividir a 'a' = "))
    a = a/b

except ZeroDivisionError:
    print("No es posible realizar la divisi贸n ya que usted ingres贸 a = 0")
except ValueError:
    print("Usted no ingres贸 n煤mero alguno.")
except:
    print("Algo sali贸 mal, que puede ser?")
else:
    print("Resultado =",a)
finally:
    print("Felicidades, pudo hacer la divisi贸n")
```

```python
# Ejempo de manejo de excepciones/errores dentro de una funci贸n

def operacion_compleja(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        print("Error: Divisi贸n por cero no permitida")
    else:
        print(f"Resultado de la operaci贸n: {resultado}")
    finally:
        print("Este bloque siempre se ejecutar谩")

# Ejemplo de uso
operacion_compleja(8, 4)
operacion_compleja(6, 0)
    
```

En el contexto de la programaci贸n orientada a objetos (OOP), una clase es una plantilla o un "molde" que define la estructura y el comportamiento de objetos. Los objetos son instancias de clases, y las clases proporcionan un mecanismo para organizar y encapsular el c贸digo de manera que sea reutilizable y f谩cil de entender.

**Principales Caracter铆sticas de las Clases:**

* **Atributos:** Las clases pueden tener atributos que representan caracter铆sticas o propiedades del objeto. Estos atributos pueden ser variables que almacenan datos.

* **M茅todos**: Las clases tambi茅n pueden contener m茅todos, que son funciones asociadas a la clase y que definen el comportamiento del objeto. Los m茅todos operan sobre los atributos del objeto.

* **Encapsulaci贸n**: Las clases permiten encapsular la implementaci贸n interna, ocultando detalles complejos y exponiendo solo la interfaz necesaria para interactuar con el objeto. Esto facilita la modularidad y el mantenimiento del c贸digo.

* **Herencia**: La herencia es un concepto clave en OOP que permite que una clase herede atributos y m茅todos de otra clase. Esto fomenta la reutilizaci贸n del c贸digo y la creaci贸n de jerarqu铆as de clases.

* **Polimorfismo**: El polimorfismo permite que objetos de diferentes clases respondan de manera consistente a un mismo conjunto de m茅todos. Esto facilita el uso de interfaces comunes en situaciones diversas.

## 3.2.1 Encapsulamiento

Permite ocultar los detalles internos de un objeto. Se logra usando atributos privados (`_` o `__`).

---
Encapsulamiento

---

La encapsulaci贸n es una de las caracter铆sticas clave de la programaci贸n orientada a objetos. La encapsulaci贸n consiste en agrupar atributos y m茅todos dentro de una misma clase. Evita que las clases externas accedan a los atributos y m茅todos de una clase y los modifiquen. Esto tambi茅n ayuda a lograr la ocultaci贸n de datos. En Python, denotamos atributos privados utilizando el gui贸n bajo como prefijo, es decir,` _` o `__`.

* `__init__`:
  * Es un m茅todo especial en Python que se llama autom谩ticamente cuando se crea una nueva instancia de una clase.
  * Este m茅todo se utiliza para inicializar los atributos de la clase o realizar otras acciones de configuraci贸n necesarias al crear un objeto.
  * Se usa para asignar valores iniciales a los atributos de la clase y puede aceptar argumentos que se pasan al crear una instancia.

* `__variable`:

  * La convenci贸n de nombres con doble guion bajo (`__`) se utiliza para indicar que una variable es privada y no debe ser accedida directamente desde fuera de la clase.
  * Aunque en Python no hay restricciones estrictas de acceso, la convenci贸n `__variable` sugiere que la variable es para uso interno y no debe ser modificada externamente.
  * La encapsulaci贸n en Python se basa en esta convenci贸n y la buena pr谩ctica de no acceder directamente a variables privadas desde fuera de la clase.

La definici贸n de una clase con encapsulaci贸n implica el uso de palabras clave tales como `class` y `__init__()`, veamos unos ejemplos.

```python
# Sintaxis b谩sica de una clase

class un_tipo_de_objeto:
  def __init__(parametros o atributos):        # Esta l铆nea de c贸digo inicializa la clase
    self.atributo_intr铆nseco_al_objeto_1 = valor_1
    self.atributo_intr铆nseco_al_objeto_2 = valor_2

  def metodo_del_objeto(parametros):
    expresion
```

En Python, la palabra reservada `self` se utiliza como convenci贸n para referirse al objeto actual de una clase. En el contexto de las clases y objetos, `self `se utiliza como el primer par谩metro de los m茅todos de una clase y representa la instancia de la propia clase.

```python
# Ejemplo de definici贸n de una clase con encapsulamiento

class Persona:
    def __init__(self, nombre, edad):
        self.__nombre = nombre  # Atributo privado
        self.__edad = edad      # Atributo privado

    def obtener_nombre(self):
        return self.__nombre

    def obtener_edad(self):
        return self.__edad

    def imprimir_informacion(self):
        print(f"Nombre: {self.__nombre}, Edad: {self.__edad}")

# Crear una instancia de la clase
persona1 = Persona("Alice", 25)

# Intentar acceder directamente a los atributos privados (no es recomendado)
# print(persona1.__nombre)  # Esto generar铆a un error

# Acceder a trav茅s de m茅todos
print(f"Nombre: {persona1.obtener_nombre()}, Edad: {persona1.obtener_edad()}")

# Imprimir informaci贸n utilizando un m茅todo p煤blico
persona1.imprimir_informacion()
```

```python
class CuentaBancaria:
    def __init__(self, saldo):
        self.__saldo = saldo  # atributo privado

    def depositar(self, monto):
        self.__saldo += monto

    def ver_saldo(self):
        return self.__saldo

cuenta = CuentaBancaria(1000)
cuenta.depositar(500)
print("Saldo actual:", cuenta.ver_saldo())
```

```python
# Ejemplo de definici贸n de una clase con encapsulamiento

class Computador:

    def __init__(self):
        self.__preciomaximo = 900

    def venta(self):
        print("Precio de venta: {}".format(self.__preciomaximo))

    def FijarPrecioMaximo(self, precio):
        self.__preciomaximo = precio

c = Computador()
c.venta()

# Intento de cambio de precio
c.__preciomaximo = 1000
c.venta()

# Usar una funci贸n que tome como argumento el atributo privado para cambiar el atributo privado
c.FijarPrecioMaximo(1000)
c.venta()
```

```python
# Ejemplo de una clase como un objeto usado para procesar datos

class ProcesadorNumeros:
    def __init__(self, numeros):
        self.numeros = numeros

    def sumar_elementos(self):
        suma = 0
        for numero in self.numeros:
            suma += numero
        return suma

# Crear una instancia de la clase
numeros_a_procesar = [1, 2, 3, 4, 5]
procesador = ProcesadorNumeros(numeros_a_procesar)

# Llamar al m茅todo para sumar elementos
resultado_suma = procesador.sumar_elementos()

# Imprimir el resultado
print(f"La suma de los elementos es: {resultado_suma}")
```

```python
# Ejemplo de una clase Calculadora con encapsulamiento

class Calculadora:
    def __init__(self, numero1, numero2):
        self.__numero1 = numero1
        self.__numero2 = numero2

    def sumar(self):
        return self.__numero1 + self.__numero2

    def restar(self):
        return self.__numero1 - self.__numero2

    def multiplicar(self):
        return self.__numero1 * self.__numero2

    def dividir(self):
        if self.__numero2 != 0:
            return self.__numero1 / self.__numero2
        else:
            print("Error: Divisi贸n por cero.")
            return None

# Crear una instancia de la clase
calculadora = Calculadora(10, 5)

# Llamar a los m茅todos encapsulados
resultado_suma = calculadora.sumar()
resultado_resta = calculadora.restar()
resultado_multiplicacion = calculadora.multiplicar()
resultado_division = calculadora.dividir()

# Imprimir resultados
print(f"Suma: {resultado_suma}")
print(f"Resta: {resultado_resta}")
print(f"Multiplicaci贸n: {resultado_multiplicacion}")
print(f"Divisi贸n: {resultado_division}")
```

## 3.2.2 Herencia

Una clase puede heredar atributos y m茅todos de otra.

---
Herencia

---

La herencia es un mecanismo en programaci贸n orientada a objetos donde una clase puede adquirir atributos y m茅todos de otra clase. La clase derivada hereda las caracter铆sticas de la clase base, lo que facilita la reutilizaci贸n del c贸digo y la creaci贸n de una jerarqu铆a de clases.

```python
# Sintaxis b谩sica de herencia de clases

class clase_1:
  atributo_1 = valor
  atributo_2 = valor

   def metodo_1(self):
    expresion

  def metodo_2(self):
    expresion

class clase_2(clase_2): # Clase que hereda los metodos de otra clase
  atributo_1 = valore
  atributo_2 = valor

  def metodo_1(self):
    expresion
```

```python
# Ejemplo de herencia simple

class persona:

  def nombre(self):
    print('Mario')

  def apellido(self):
    print('Rodriguez')

class otra_persona(persona):
  def comentario(self):
    print('Puedo hacer comentarios!')

p = otra_persona()

p.nombre()
p.apellido()
p.comentario()
```

```python
# Ejemplo de clase cveh铆culo

class Vehiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def mostrar_informacion(self):
        print(f"Marca: {self.marca}, Modelo: {self.modelo}")

class Coche(Vehiculo):
    def __init__(self, marca, modelo, color):
        # Llamada al constructor de la clase base
        super().__init__(marca, modelo)
        self.color = color

    def mostrar_informacion_coche(self):
        print(f"Color: {self.color}")

# Crear una instancia de la clase derivada (Coche)
mi_coche = Coche("Toyota", "Camry", "Azul")

# Acceder a los m茅todos de la clase base y la clase derivada
mi_coche.mostrar_informacion()  # M茅todos de la clase base
mi_coche.mostrar_informacion_coche()  # M茅todo de la clase derivada
```

## 3.2.3 Polimorfismo

Permite que diferentes clases compartan el mismo m茅todo con distintas implementaciones.

```python
class Gato(Animal):
    def hablar(self):
        print("El gato dice: Miau")

animales = [Perro("Toby"), Gato("Michi")]

for animal in animales:
    animal.hablar()  # comportamiento diferente seg煤n el tipo
```

#  3.3 M茅todos especiales

## 3.3.1 M茅todos especiales (`__init__`, `__str__`, `__repr__`, etc.)

Estos m茅todos permiten personalizar el comportamiento de los objetos en situaciones espec铆ficas.

```python
class Libro:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor

    def __str__(self):
        return f"{self.titulo} por {self.autor}"

    def __repr__(self):
        return f"Libro('{self.titulo}', '{self.autor}')"

libro = Libro("1984", "George Orwell")
print(str(libro))
print(repr(libro))
```

## 3.3.2 Sobrecarga de operadores (`__add__`, `__eq__`, etc.)

Permite redefinir c贸mo se comportan los operadores (`+`, `==`, etc.) entre objetos personalizados.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, otro):
        return Vector(self.x + otro.x, self.y + otro.y)

    def __eq__(self, otro):
        return self.x == otro.x and self.y == otro.y

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2

print(v3)
print("驴v1 y v2 son iguales?", v1 == v2)
```
